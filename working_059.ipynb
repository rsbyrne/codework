```python
ang, rad = pfn.box.getall(system.mesh)
pert = 0.2
phase = 0.
freq = 10.
pertFn = \
    pert \
    * fn.math.cos(3.14159 * (phase + freq * ang)) \
    * fn.math.sin(3.14159 * rad)
sinuFn = rad + pertFn
```

```python
quickShow(sinuFn)
```
