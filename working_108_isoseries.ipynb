```python
import planetengine
```

```python
System = planetengine.systems.Isovisc
sE4 = System(res = 16, alpha = 1e4)
sE5 = System(res = 32, alpha = 1e5, temperatureField = sE4[:300])
sE6 = System(res = 64, alpha = 1e6, temperatureField = sE5[:500])
system = System(res = 64, alpha = 1e7, temperatureField = sE6[:800])
```

```python
system.anchor('isotest', '.')
```

```python
system.initialise()
```

```python
system.load('max')
```

```python
system.add_observer()
```

```python
system.show()
```
