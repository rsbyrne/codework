```python
from planetengine.functions import conduction
from planetengine import quickShow
from planetengine import systems
system = systems.isovisc(f = 0.5)
```

```python
HField = system.locals.mesh.add_variable(1)
HField.data[:] = 10.
```

```python
cond = conduction.inner(HField, 1., 10.)
quickShow(cond)
```
