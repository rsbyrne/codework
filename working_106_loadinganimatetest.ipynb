```python
from planetengine.systems import Isovisc as System
from planetengine.initials import Copy
system1 = System(
    res = 16,
    alpha = 1e5,
    f = 0.5,
    aspect = 1.414
    )
system2 = System(
    res = 32,
    alpha = 1e6,
    f = 0.5,
    aspect = 1.414,
    temperatureField = Copy(system1[:1000], 'temperatureField')
    )
system3 = System(
    res = 64,
    alpha = 1e7,
    f = 0.5,
    aspect = 1.414,
    temperatureField = Copy(system2[:300], 'temperatureField')
    )
system4 = System(
    res = 64,
    alpha = 1e7,
    f = 0.5,
    aspect = 'max',
    temperatureField = Copy(system3[:100], 'temperatureField', mirrored = (True, False), tiles = (3., 1.))
    )
```

```python
system4.add_observer()
```

```python
system4.report()
```

```python
system4.show()
```

```python
traverse = system4[::100]
```

```python
traverse()
```
