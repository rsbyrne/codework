```python
from PIL import Image
import numpy as np
from IPython.display import display

import planetengine
pfn = planetengine.functions
quickShow = planetengine.visualisation.quickShow
```

```python
system = planetengine.tests.testsystems.isovisc(res = 32, Ra = 1e6)
```

```python
channel0 = pfn.component.mag(system.stokes.fn_bodyforce)
channel1 = pfn.component.mag(system.velocityField)
channel2 = pfn.component.rad(system.velocityField)
```

```python
quickShow(channel0)
my_raster = planetengine.raster.Raster(channel0, add = system.step)
my_raster.img.resize((256, 256))
```

```python
quickShow(channel1)
my_raster = planetengine.raster.Raster(channel1)
my_raster.img.resize((256, 256))
```

```python
quickShow(channel2)
my_raster = planetengine.raster.Raster(channel2)
my_raster.img.resize((256, 256))
```

```python
my_raster = planetengine.raster.Raster(channel0, channel1)
my_raster.img.resize((256, 256))
```

```python
my_raster = planetengine.raster.Raster(channel0, channel1, channel2)
my_raster.img.resize((256, 256))
```

```python
my_raster = planetengine.raster.Raster(channel0, channel1, channel2)
my_raster.update()
myfig = planetengine.visualisation.QuickFig(system.temperatureField)
display(my_raster.img.resize((256, 256)))
system.reset()
for i in range(1000):
    system.go(10)
    myfig.update()
#     myfig.show()
    myfig.save('../testrasters', 'testfig_' + str(system.step.value).zfill(8) + '.png')
    my_raster.update()
#     display(my_raster.img.resize((256, 256)))
    my_raster.img.save('../testrasters/testraster_' + str(system.step.value).zfill(8) + '.png')
```
