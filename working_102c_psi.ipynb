```python
from planetengine import systems
```

```python
system = systems.viscoplastic(f = 0.5)
```

```python
from planetengine.functions import stream
```

```python
psi = stream.default(system.locals.vc)
```

```python
from planetengine import quickShow
```

```python
quickShow(system.locals.temperatureField, psi, system.locals.velocityField)
```

```python
from planetengine.functions import gradient, component
```

```python
psix, psiy = gradient.x(psi), gradient.y(psi)
```

```python
u, v = component.x(system.locals.vc), component.y(system.locals.vc)
```

```python
xv = psix / -v
yu = psiy / u
```

```python
quickShow(xv)
```

```python
quickShow(yu)
```

```python
yu.evaluate()
```

```python
max(xv.evaluate())
```
