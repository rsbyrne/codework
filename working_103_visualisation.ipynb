```python
from planetengine.visualisation import raster
from IPython.display import Video
from planetengine import systems
```

```python
system = systems.isovisc(res = 32, alpha = 1e5, aspect = 3.)
system.add_observer()
```

```python
system.observer.raster.img
```

```python
for i in range(100):
    system.iterate(20)
    system.observer.store()
filePath = raster.animate(system.observer.make_dataDict()['raster'])
Video(filePath, embed = True)
```

```python
system.show()
```
